```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import healpy as hp
import os
catalog_list = os.listdir('./data/unWISE_catalog/')
```

```python
NSIDE = 2048

from astropy.coordinates import SkyCoord
from astropy import units as u


for i in range(len(catalog_list)):
    with fits.open('./data/unWISE_catalog/'+catalog_list[i]) as hdul:
        data = hdul[1].data

    ra = data['ra']
    dec = data['dec']

    coords = SkyCoord(ra=ra*u.degree, dec=dec*u.degree, frame='icrs')
    l,b = coords.galactic.l.value, coords.galactic.b.value

    flux_w1 = data['flux_w1']
    flux_w2 = data['flux_w2']

    dflux_w1 = data['dflux_w1']
    dflux_w2 = data['dflux_w2']

    pix = hp.ang2pix(NSIDE, l, b,lonlat=True)
    map_w1 = np.bincount(pix, weights=flux_w1, minlength=hp.nside2npix(NSIDE))
    map_w2 = np.bincount(pix, weights=flux_w2, minlength=hp.nside2npix(NSIDE))
    map_w1_d = np.bincount(pix, weights=dflux_w1**2, minlength=hp.nside2npix(NSIDE))
    map_w2_d = np.bincount(pix, weights=dflux_w2**2, minlength=hp.nside2npix(NSIDE))
    map_count = np.bincount(pix, minlength=hp.nside2npix(NSIDE))

    np.save('./data/unWISE_map/map_w1_{}.npy'.format(i),map_w1)
    np.save('./data/unWISE_map/map_w2_{}.npy'.format(i),map_w2)
    np.save('./data/unWISE_map/map_w1_d_{}.npy'.format(i),map_w1_d)
    np.save('./data/unWISE_map/map_w2_d_{}.npy'.format(i),map_w2_d)
    np.save('./data/unWISE_map/map_count_{}.npy'.format(i),map_count)
    print('done mapping '+catalog_list[i])



    
```

```python
mapw1 = np.zeros(hp.nside2npix(NSIDE))
mapw1d = np.zeros_like(map_w1)
mapw2 = np.zeros_like(map_w1)
mapw2d = np.zeros_like(map_w1)
mapcount = np.zeros_like(map_w1)

for i in range(10):
    mapw1 += np.load('./data/unWISE_map/map_w1_{}.npy'.format(i))
    mapw2 += np.load('./data/unWISE_map/map_w2_{}.npy'.format(i))
    mapw1d += np.load('./data/unWISE_map/map_w1_d_{}.npy'.format(i))
    mapw2d += np.load('./data/unWISE_map/map_w2_d_{}.npy'.format(i))
    mapcount += np.load('./data/unWISE_map/map_count_{}.npy'.format(i))

mapw1 /= mapcount
mapw2 /= mapcount
mapw1d /= mapcount
mapw2d /= mapcount

mapw1d = np.sqrt(mapw1d)
mapw2d = np.sqrt(mapw2d)
```

```python
mapw1[np.isnan(mapw1)] = 0
mapw2[np.isnan(mapw2)] = 0
mapw1d[np.isnan(mapw1d)] = 0
mapw2d[np.isnan(mapw2d)] = 0
```

```python
np.save('./data/unWISE_result/map_w1.npy',mapw1)
np.save('./data/unWISE_result/map_w2.npy',mapw2)
np.save('./data/unWISE_result/map_w1_d.npy',mapw1d)
np.save('./data/unWISE_result/map_w2_d.npy',mapw2d)
np.save('./data/unWISE_result/map_count.npy',mapcount)
```
